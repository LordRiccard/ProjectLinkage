```python
import glob
import pandas as pd
```

## SIHSUS

```python
directory = 'D:\\_repositories\\Databases\\SIHSUS'
pattern = f'*.csv'

files = glob.glob(f'{directory}/{pattern}')
print(files)
```

```python
lista = []

for file in files:
    try:
        df = pd.read_csv(file)
    except UnicodeDecodeError:
        lista.append(file)
```

```python
print(lista)
```

```python
for item in lista:
    df = pd.read_csv(item, encoding='ANSI')
    df = df.drop_duplicates()

    lista_colunas = df.columns
    
    for coluna in lista_colunas:
        result = df.index[df[coluna] == coluna].tolist()
    #print(result)
    df = df.drop(index=result)

    df.to_csv('D:\\_repositories\\Aggregation\\Correo de Erros\\Output\\SIHSUS\\' + item[34:], encoding='UTF-8')
```

## SINASC

```python
directory = 'D:\\_repositories\\Databases\\SINASC\\'
pattern = f'*\\*.csv'

files = glob.glob(f'{directory}/{pattern}')
print(files)
```

```python
def find_double_header(df):
    lista_colunas = df.columns
    
    for coluna in lista_colunas:
        result = df.index[df[coluna] == coluna].tolist()
        df = df.drop(result)

    return df
```

```python
for item in files:
    df = pd.read_csv(item)

    df = df.drop_duplicates() # it is not going to find the duplicates
    df = find_double_header(df)

    df.to_csv('D:\\_repositories\\Aggregation\\Results\\SINASC' + item[36:], index=False)
```
