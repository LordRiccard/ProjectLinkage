## Tentativa de Agregação do SIHSUS

```python
def get_files_ano(i, est):
    master = []

    for j in range(1, 13):
        directory = 'D:\\_repositories\\Databases\\SIHSUS'

        if j < 10:
            pattern = f'RD{est}*{i}0{j}.csv'
        else:
            pattern = f'RD{est}*{i}{j}.csv'

        files = glob.glob(f'{directory}/{pattern}')
        # print(files)
        for item in files:
            master.append(item)


    

    return master
# backed
```

```python
get_files_ano(22, 'SP')
```

```python
import warnings
warnings.filterwarnings('ignore', category=pd.errors.DtypeWarning)
```

```python
estados_brasil = ["AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO", "MA", "MT", "MS", "MG", "PA", "PB", "PR", "PE", "PI", "RJ", "RN", "RS", "RO", "RR", "SC", "SP", "SE", "TO"]

for est in estados_brasil:
    for i in range (12, 23):

        master = get_files_ano(i, est)

        aux = pd.DataFrame()

        for item in master:
            df = pd.read_csv(item)
            aux = pd.concat([aux, df], axis=0)
            aux = aux.drop_duplicates()

        aux.to_csv(f'D:\\_repositories\\Aggregation\\Results\\SIH - Meses Juntos\\RD{est}20{i}.csv', index=False)
        print(f'finalizado {est} do ano 20{i}')
        
```
